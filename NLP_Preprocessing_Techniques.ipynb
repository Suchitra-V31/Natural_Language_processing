Let us use these some of the preprocessing techniques to solve the gutenberg data.

**PREPROCESSING**

```python
pre_sent=[]
for s in gberg_sent:
 pre_sent.append([w.lower() for w in s if w.lower()  if s not in list(string.punctuation)]) 
```

```python
pre_sent[0]
```

```python
phrases=Phrases(pre_sent)
bigram=Phraser(phrases)
```

```python
bigram.phrasegrams
```

```python
bigram=Phraser(Phrases(pre_sent,min_count=32,threshold=64))
```

```python
bigram.phrasegrams
```

```python
final_sent=[]
for s in pre_sent:
  final_sent.append(bigram[s])
```

```python
final_sent[6]
```

**Run word2vec**

```python
model=Word2Vec(sentences= final_sent,size=64,window=8,min_count=5,iter=4,sg=1)
```

```python
model.wv['dog']
```

```python
len(model.wv['dog'])
```

```python
model.wv.most_similar('dog',topn=3)
```

```python
model.wv.doesnt_match("mother father sister brother dog".split())
```

```python
model.wv.similarity('father', 'dog')
```

```python
model.wv.most_similar(positive=['father', 'woman'], negative=['man'])
```

**Reduce dimension using t-SNE**

```python
 tsne = TSNE(n_components=2, n_iter=1000)
 X_2d = tsne.fit_transform(model.wv[model.wv.vocab])
```

```python
df = pd.DataFrame(X_2d, columns=['x','y'])
df['token'] = model.wv.vocab.keys()
```

```python
df.head()
```

**Visualize text**

```python
output_notebook()
```

```python
sample_df=df.sample(n=5000)
```

```python
p=figure(plot_width=800,plot_height=800)
_=p.text(x=sample_df.x,y=sample_df.y,text=sample_df.token)
```
